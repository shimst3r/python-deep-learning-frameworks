# Development over Time

This notebook plots the development over time for each repository, split into the following comparisons:

- `open_issues` per `stargazers`
- `forks` per `open_issues`
- `stargazers` over `timestamp`
- `open_issues` over `timestamp`

```python
import pandas as pd
import seaborn as sns; sns.set()
```

```python
df = pd.read_csv('../repositories.csv', delimiter='\t')
```

```python
ax = sns.scatterplot(
    x='stargazers',
    y='open_issues',
    hue='name',
    data=df[df.stargazers > df.stargazers.quantile(0.75)].sort_values(by=['stargazers'], ascending=False),
)
ax.figure.set_figwidth(16)
ax.figure.set_figheight(9)
sns.move_legend(ax, 'upper left', bbox_to_anchor=(1, 1))
```

```python
ax = sns.scatterplot(
    x='open_issues',
    y='forks',
    hue='name',
    data=df[df.stargazers > df.stargazers.quantile(0.75)].sort_values(by=['stargazers'], ascending=False),
)
ax.figure.set_figwidth(16)
ax.figure.set_figheight(9)
sns.move_legend(ax, 'upper left', bbox_to_anchor=(1, 1))
```

```python
ax = sns.lineplot(
    x='timestamp',
    y='stargazers',
    hue='name',
    data=df[df.stargazers > df.stargazers.quantile(0.75)].sort_values(by=['timestamp'], ascending=True),
)
ax.figure.set_figwidth(16)
ax.figure.set_figheight(9)
sns.move_legend(ax, 'upper left', bbox_to_anchor=(1, 1))
```

```python
ax = sns.lineplot(
    x='timestamp',
    y='open_issues',
    hue='name',
    data=df[df.stargazers > df.stargazers.quantile(0.75)].sort_values(by=['timestamp'], ascending=True),
)
ax.figure.set_figwidth(16)
ax.figure.set_figheight(9)
sns.move_legend(ax, 'upper left', bbox_to_anchor=(1, 1))
```
